```python
# Filtering Rows Based on Conditions-------------------------

"""
Filtering data in pandas is one of the most common tasks when working with DataFrames.
    It allows you to extract rows or columns that meet specific conditions.
    Below are the most common methods for filtering data in pandas:

1. Filtering Rows Based on Conditions
  You can filter rows by applying conditions to one or more columns.
  The result is a DataFrame containing only the rows that satisfy the condition(s).

"""
condition1=df[df['QUANTITYORDERED'] > 50]
condition2=df[(df['QUANTITYORDERED'] > 50) & (df['STATUS'] == 'Shipped')]
condition3=df[(df['QUANTITYORDERED'] > 50) | (df['STATUS'] == 'Shipped')]
condition4=df[~(df['QUANTITYORDERED'] > 50)]
# print(condition1.head())
# print(condition2.head())
# print(condition3.head())
# print(condition4.head())


"""

3. Filtering with .query()
  The .query() method allows you to filter rows using a string expression.

"""
filtered_df=df.query('QUANTITYORDERED > 50')
filtered_df=df.query('QUANTITYORDERED > 50 and STATUS == "Shipped"')
filtered_df=df.query('QUANTITYORDERED > 50 or STATUS == "Shipped"')
filtered_df=df.query('not QUANTITYORDERED > 50')
# print(filtered_df.head())

"""

4. Filtering with .isin()
The .isin() method is useful for filtering rows where a column's value is in a list of values.

"""
filtered_df=df[df['STATUS'].isin(['Shipped','Cancelled'])]
filtered_df=df[~df['STATUS'].isin(['Shipped','Cancelled'])]
filtered_df=df.query('STATUS in ["Shipped","Cancelled"]')
filtered_df=df.query('STATUS not in ["Shipped","Cancelled"]')
# print(filtered_df.head())


"""

5. Filtering with .str.contains()
For string columns, you can filter rows where a column contains a specific substring.

"""
filtered_df=df[df['STATUS'].str.contains('Ship')]
filtered_df=df[~df['STATUS'].str.contains('Ship')]
filtered_df=df.query('STATUS.str.contains("Ship")')
filtered_df=df.query('not STATUS.str.contains("Ship")')
# print(filtered_df.head())

# Filtering with .str.startswith() and .str.endswith()
filtered_df=df[df['STATUS'].str.startswith('Ship')]
filtered_df=df[df['STATUS'].str.endswith('ed')]
# print(filtered_df.head())

"""

6. Filtering with .loc[]
The .loc[] method allows you to filter rows and select specific columns simultaneously.

"""
filtered_df=df.loc[df['QUANTITYORDERED'] > 50,['CITY']]
filtered_df=df.loc[(df['QUANTITYORDERED'] > 50) & (df['STATUS'] == 'Shipped'),['CITY']]
# print(filtered_df.head())


"""

8. Filtering Missing Values
You can filter rows based on missing or non-missing values using .isna() or .notna().

"""
filtered_df=df[df['POSTALCODE'].isna()] # Filter rows where 'POSTALCODE' is  missing
filtered_df=df[df['POSTALCODE'].notna()] ## Filter rows where 'POSTALCODE' is not missing
# print(filtered_df.head())


"""

9. Filtering with .between()
The .between() method is useful for filtering rows where a column's value falls within a range.

"""
filtered_df=df[df['QUANTITYORDERED'].between(20,25)]
filtered_df=df[(df['QUANTITYORDERED'] >= 20) & (df['QUANTITYORDERED'] <= 25)]
# print(filtered_df.head())


"""

10. Filtering with .apply()
You can use .apply() to filter rows based on a custom function.

"""
def custom_filter(row):
    return row['QUANTITYORDERED'] > 50 and row['STATUS'] == 'Shipped'

filtered_df=df[df.apply(custom_filter, axis=1)]
#note-(axis=1 ensures the function is applied row-wise.)
#note-( custom_filter function to each row of the DataFrame.)

filtered_df=df[df['CITY'].apply(lambda x : x.startswith('M'))] # using lambda funciton ----------------
print(filtered_df.head())



"""

Summary of Filtering Methods:
Method	                            Use Case
df[df['column'] > value]	          Filter rows based on a condition.
&, `	, ~`	                        Combine multiple conditions.
.query()	                          Filter rows using a string expression.
.isin()	                            Filter rows where a column's value is in a list.
.str.contains()	                    Filter rows where a string column contains a substring.
.loc[]	                            Filter rows and select columns by labels.
.iloc[]	                            Filter rows and select columns by integer positions.
.isna(), .notna()	                  Filter rows based on missing or non-missing values.
.between()	                        Filter rows where a column's value is within a range.
.apply()	                          Filter rows using a custom function.
.str.startswith()                   Filter rows where a string column starts with a substring.
.str.endswith()                     Filter rows where a string column ends with a substring.
"""
```

```python
# Grouping

"""

Grouping in Pandas: Concepts and Problem Set
      Grouping is one of the most powerful features in pandas,
      allowing you to split your data into groups, apply operations
      to each group, and combine the results. Let's explore this
      with a comprehensive problem set.


"""
# Basic Grouping Concepts
# 1. groupby() Fundamentals

import pandas as pd

# Sample DataFrame
data = {
    'Department': ['HR', 'IT', 'Finance', 'HR', 'IT', 'Finance'],
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank'],
    'Salary': [70000, 80000, 90000, 75000, 85000, 95000],
    'Experience': [3, 5, 7, 4, 6, 8]
}

df=pd.DataFrame(data)

grouped=df.groupby('Department')
grouped.first()
grouped.get_group('HR')
grouped.get_group('IT')
grouped.get_group('Finance')
# grouped.head()
print(grouped.groups)
    # 'Finance': [2, 5],  # Rows 2 and 5 (0-based index)
    # 'HR': [0, 3],       # Rows 0 and 3
    # 'IT': [1, 4]        # Rows 1 and 4


# 2. Common Aggregation Operations

# Single aggregation
df.groupby('Department')['Salary'].mean()

# Multiple aggregations
df.groupby('Department').agg({
    'Salary': ['mean', 'min', 'max', 'std'],
    'Experience': 'median'
})

# 3. Grouping by Multiple Columns

# Adding a 'Location' column
df['Location'] = ['NY', 'SF', 'NY', 'SF', 'NY', 'SF']
# Group by multiple columns
df.groupby(['Department', 'Location'])['Salary'].mean()
```

```python
# Problem Set
# Problem 1: Basic Grouping

#1. Calculate total revenue by region

#2. Find average units sold by product

#3. Get both total revenue and average units by region and product
import pandas as pd

sales = pd.DataFrame({
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
    'Product': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Revenue': [100, 150, 200, 120, 180, 220, 190, 130],
    'Units': [10, 15, 20, 12, 18, 22, 19, 13]
})
#1
sales.groupby('Region')['Revenue'].sum()
#2
sales.groupby('Product')['Units'].mean()
#3
sales.groupby(['Region','Product']).agg({
    'Revenue': 'sum',
    'Units': 'mean'
})



# Problem 2: Advanced Aggregations
# Using the same sales data:

# 1. Calculate the percentage each product contributes to total revenue in each region

# 2. Find the difference between maximum and minimum revenue for each product

# 3. Create a custom aggregation that calculates revenue per unit

#1
region_product=sales.groupby(['Region','Product'])['Revenue'].sum()
region_total=sales.groupby('Region')['Revenue'].sum()
print((region_product/region_total*100).round(2))

#2
sales.groupby('Product').agg({
    'Revenue': lambda x: x.max()-x.min()
})

sales.groupby('Product').agg({
    'Revenue': lambda x: x.sum()/x.count()
})

#3
def revenue_per_unit(group):
    return group['Revenue'].sum() / group['Units'].sum()

sales.groupby(['Region', 'Product']).apply(revenue_per_unit)
sales['Revenue_per_unit']=sales.groupby(['Product','Region'])['Revenue'].transform(lambda x: x.sum()/sales.groupby(['Product','Region'])['Units'].transform('sum'))
sales.head()



# Problem 3: Time-Based Grouping

dates = pd.date_range('2023-01-01', periods=12, freq='M')
time_sales = pd.DataFrame({
    'Date': dates,
    'Product': ['A', 'B'] * 6,
    'Sales': [100, 150, 120, 180, 130, 200, 140, 170, 160, 190, 200, 210]
})

time_sales['Quarter']=time_sales['Date'].dt.quarter
time_sales.groupby(['Quarter','Product'])['Sales'].sum()


time_sales['Monthtly_sales_difference']=time_sales.groupby('Product')['Sales'].diff()
time_sales.head()

q1_q2=time_sales[time_sales['Quarter'].isin([1,2])]
q1_q2.groupby(['Quarter','Product'])['Sales'].sum().unstack()
```

```python
# Problem 4: Complex Grouping Operations

#1. Calculate average salary by department and gender

#2. Find the age difference between oldest and youngest employee in each department

#3. Create a salary percentile rank within each department

employees = pd.DataFrame({
    'Department': ['HR', 'IT', 'Finance', 'HR', 'IT', 'Finance', 'HR', 'IT'],
    'Gender': ['F', 'M', 'M', 'F', 'M', 'F', 'F', 'M'],
    'Salary': [70000, 80000, 90000, 75000, 85000, 95000, 78000, 82000],
    'Age': [32, 28, 45, 29, 31, 38, 35, 27]
})

employees.groupby(['Department','Gender'])['Salary'].mean()

employees.groupby('Department')['Age'].apply(lambda x: x.max()-x.min())

employees['Dept_salary_rank']=employees.groupby('Department')['Salary'].rank(ascending=False,pct=True)
employees.sort_values(['Department','Dept_salary_rank'])


# Key Takeaways
      # groupby() splits data into groups based on specified criteria

      # You can apply various aggregation functions to grouped data

      # Multiple grouping keys can be used together

      # Custom aggregation functions provide flexibility

      # Time-based grouping is powerful for temporal analysis

      # Operations within groups (like ranking) are often useful
```

```python
# Sorting in Pandas: A Comprehensive Guide
# Pandas provides powerful sorting capabilities for both Series and DataFrames. Here's a complete guide to sorting with practical examples.

# Basic Sorting Methods
# 1. Sorting by Values
import pandas as pd

# Create a sample DataFrame
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 22, 35],
    'Salary': [50000, 70000, 45000, 80000]
})
# Sort by a single column (ascending)
df_sorted = df.sort_values('Age')
df_sorted.head()

# Sort descending
df_sorted=df.sort_values('Age', ascending=False)
df_sorted.head()

# Sort by multiple columns
df_sorted_multi=df.sort_values(['Age','Salary'],ascending=[True,False])
df_sorted_multi.head()
```

```python
# Handling Missing Values
# 1. Dropping missing values:
data = {
    'A': [1, 2, None, 4],
    'B': [2.5, None, 5.5, 6.0],
    'C': ['x', 'y', None, 'z'],
    'D': [pd.Timestamp('20200101'), None, None, pd.Timestamp('20200104')]
}
df = pd.DataFrame(data)

df.dropna() # Drop rows with any missing values
df.dropna(axis=1)  # Drop columns with any missing values
df.dropna(thresh=3)  # Keep rows with at least 3 non-NA values


# 2. Filling missing values:
df.fillna(0)  # Fill with a scalar value
df.fillna(method='ffill')  # Forward fill (fill with previous value)
df.fillna(method='bfill')  # Backward fill (fill with next value)
df.fillna({'A':0,'B':df['B'].mean()}) # Different fills per column
df.ffill() # Forward fill
df.bfill() # Backward fill


# 3. Interpolation:
df.interpolate() # Linear interpolation
# df.interpolate(method='polynomial', order=2) # Polynomial interpolation
```

```python
"""
Real-World Problem: Time Series Data with Gaps
Scenario: You're analyzing daily temperature data with some days
missing due to sensor failures.
Problem: Missing values break time series analysis and visualization.

Solution Options:

"""

# Option 1: Forward fill (carry last observation forward)
df['temperature'].ffill(inplace=True)

# Option 2: Interpolate based on time
df['temperature'].interpolate(method='time', inplace=True)

# Option 3: Fill with seasonal average
df['temperature'].fillna(df.groupby('month')['temperature'].transform('mean'), inplace=True)

```

```python
# 3. Interaction Between dtype and Missing Values

# a.  Type-Specific Missing Values

# Numeric columns get NaN
s = pd.Series([1, 2, None])
print(s.dtype)  # float64 (None becomes NaN)

# Object columns can hold None or NaN
s = pd.Series(['a', 'b', None])
print(s.dtype)  # object

# Boolean columns with None become object dtype
s = pd.Series([True, False, None])
print(s.dtype)  # object


# b.  Using pd.NA (Pandas 1.0+)

# pd.NA is a new missing value indicator that works across all dtypes:
s = pd.Series([1, 2, pd.NA], dtype='Int64')  # New nullable integer type
print(s)

s = pd.Series([True, False, pd.NA], dtype='boolean')
print(s)
```

```python
def validate_data(df):
    """Validate DataFrame structure and content"""
    # Check expected columns exist
    required_cols = ['id', 'date', 'value']
    missing = set(required_cols) - set(df.columns)
    if missing:
        raise ValueError(f"Missing columns: {missing}")

    # Check dtypes
    if not pd.api.types.is_numeric_dtype(df['value']):
        raise TypeError("'value' must be numeric")

    # Check for unexpected missing values
    if df['id'].isna().any():
        raise ValueError("Missing values in ID column")

    # Check date range
    if not pd.api.types.is_datetime64_dtype(df['date']):
        df['date'] = pd.to_datetime(df['date'], errors='coerce')
        if df['date'].isna().any():
            raise ValueError("Invalid date values")

    return True

data={
    'id':[1,2,None],
    'date':['2023-01-01','2023-01-02','2023-01-03'],
    'value':[34,53,33]
}
df=pd.DataFrame(data)
print(validate_data(df))
```

```python
# Combining DataFrames in Pandas
# Common Methods:
# 1. concat() - Stack DataFrames vertically or horizontally
# 2. merge() - SQL-style joins
# 3. join() - Convenience method for merging on indices


# Real-Life Use Case: Combining Monthly Sales Reports
# Monthly sales data for Q1
jan_sales = pd.DataFrame({
    'product_id': ['P100', 'P101', 'P102'],
    'jan_sales': [120, 95, 200]
})

feb_sales = pd.DataFrame({
    'product_id': ['P100', 'P101', 'P103'],
    'feb_sales': [150, 110, 180]
})

mar_sales = pd.DataFrame({
    'product_id': ['P100', 'P102', 'P103'],
    'mar_sales': [130, 210, 190]
})

# Combine using merge (left join to keep all products)
q1_sales=jan_sales.merge(feb_sales,on='product_id',how='left')
q1_sales=q1_sales.merge(mar_sales,on='product_id',how='left')

print("\nCombined quarterly sales:")
# print(q1_sales)

#=================================================================================

# Another Example: Combining Customer Data from Different Sources
# Customer demographic data
customers = pd.DataFrame({
    'customer_id': [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'join_date': ['2022-01-15', '2021-11-03', '2023-02-20', '2022-07-10']
})

# Customer purchase data
purchases = pd.DataFrame({
    'purchase_id': [101, 102, 103, 104],
    'cust_id': [1, 2, 1, 3],
    'amount': [49.99, 29.99, 99.99, 14.99],
    'date': ['2023-01-10', '2023-01-12', '2023-01-15', '2023-01-20']
})

# Combine with merge (note different column names for customer ID)
customer_analysis= customers.merge(purchases,left_on='customer_id',right_on='cust_id',how='left')
print("\nCustomer analysis:")
print(customer_analysis)

# Clean up by dropping the redundant column and renaming
customer_analysis=customer_analysis.drop(columns='cust_id',axis=1)
customer_analysis=customer_analysis.rename(columns={
    'date':'purchase_date'
})

print("\nCustomer analysis:")
print(customer_analysis)
```

```python
# 1. pd.concat() - Stacking DataFrames Vertically or Horizontally
# Use Cases & Examples:
# A. Vertical Concatenation (Row-wise)
# Scenario: Combining monthly sales reports into a single yearly report.


# Monthly sales data
jan_sales = pd.DataFrame({
    'product_id': ['P100', 'P101', 'P102'],
    'sales': [120, 95, 200]
})

feb_sales = pd.DataFrame({
    'product_id': ['P100', 'P103', 'P104'],
    'sales': [150, 180, 90]
})

# Combine vertically (stack rows)
yearly_sales=pd.concat([jan_sales,feb_sales],ignore_index=True)
print("\nYearly sales:")
print(yearly_sales)


# B. Horizontal Concatenation (Column-wise)
# Scenario: Adding new metrics (e.g., "profit") to an existing sales DataFrame.

sales = pd.DataFrame({
    'product_id': ['P100', 'P101', 'P102'],
    'sales': [120, 95, 200]
})

profit = pd.DataFrame({
    'profit': [30, 15, 50]
})

# Combine horizontally (side by side)
sales_report = pd.concat([sales, profit], axis=1)

print(sales_report)
```

```python
# 2. pd.merge() - SQL-Style Joins
# Used to combine datasets based on common columns (like SQL JOIN).

# Use Cases & Examples:
# A. Inner Join (Default)
# Scenario: Matching customer orders with customer details.
customers = pd.DataFrame({
    'customer_id': [1, 2, 3],
    'name': ['Alice', 'Bob', 'Charlie']
})

orders = pd.DataFrame({
    'order_id': [101, 102, 103],
    'customer_id': [1, 2, 4],
    'amount': [49.99, 29.99, 19.99]
})

# Inner join (only matching records)

merged=pd.merge(customers,orders,on='customer_id',how='inner')
print("Inner merge: ")
print(merged)


# B. Left Join (Keep All Customers)
# Scenario: Get all customers, even if they havenâ€™t placed orders.

merge_left=pd.merge(customers,orders,on='customer_id',how='left')
print("Left merge: ")
print(merge_left)

# C. Right Join (Keep All Orders)
# Scenario: Get all orders, even if customer details are missing.

merge_right=pd.merge(customers,orders,on='customer_id',how='right')
print("Right merge: ")
print(merge_right)

# D. Outer Join (Keep All Records)
# Scenario: Get all customers and all orders, including unmatched records.

merge_outer=pd.merge(customers,orders,on='customer_id',how='outer')
print("Outer merge: ")
print(merge_outer)
```

```python
# 3. df.join() - Merging on Indices
# A shortcut for merging when joining on indexes instead of columns.

# Use Cases & Examples:
# A. Joining DataFrames with Index Alignment
# Scenario: Combining stock prices (indexed by date).

prices = pd.DataFrame({
    'price': [100, 105, 110]
}, index=['2023-01-01', '2023-01-02', '2023-01-03'])

volumes = pd.DataFrame({
    'volume': [1000, 1500, 2000]
}, index=['2023-01-01', '2023-01-02', '2023-01-03'])

# Join on index
combined = prices.join(volumes)
print("Joined DataFrame:")
print(combined)


# B. Joining with Different Indexes (Left Join)
# Scenario: Merging employee data (some missing records).
employees = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie']
}, index=[1, 2, 3])

salaries = pd.DataFrame({
    'salary': [5000, 6000]
}, index=[1, 2])

# Left join (keep all employees)
left_joined=employees.join(salaries,how='left')
print("Left joined DataFrame:")
print(left_joined)



"""

Summary of Use Cases
  Method	          Best For	                                    Example Use Case
  concat()	        Stacking DataFrames (rows or columns)	        Combining monthly reports
  merge()	          SQL-style joins (inner, left, right, outer)	  Merging customer + order data
  join()	          Merging on indexes	                          Combining stock prices & volumes


"""
```

```python
# rename() in Pandas: All Use Cases with Real-World Examples
# The rename() method in Pandas is used to modify column names or row index labels in a DataFrame. It is particularly useful for:

# 1. Standardizing column names (e.g., converting to lowercase)

# 2. Making names more descriptive (e.g., amt â†’ amount)

# 3. Fixing typos (e.g., custmer_id â†’ customer_id)

# 4. Renaming index labels (e.g., changing row IDs)



# 1. Renaming Columns
# A. Basic Column Renaming (Single Column)
# Scenario: Fixing a typo in a column name.
import pandas as pd

df = pd.DataFrame({
    'custmer_id': [1, 2, 3],
    'amt': [100, 200, 300]
})

df = df.rename(columns={'custmer_id': 'customer_id'})
print(df.columns)


# B. Renaming Multiple Columns
# Scenario: Standardizing column names (e.g., converting to lowercase).
df = df.rename(columns={
    'customer_id': 'customer_id',
    'amt': 'amount'
})
print(df.columns)

# C. Renaming Columns In-Place (Without Reassignment)
# Scenario: Modifying column names directly without creating a new DataFrame.
df.rename(columns={'amount': 'sale_amount'}, inplace=True)
print(df.columns)

# D. Renaming Columns Using a Function (Lambda)
# Scenario: Applying a transformation to all column names (e.g., converting to uppercase).
df.rename(columns=lambda x: x.upper(), inplace=True)
print(df.columns)



# 2. Renaming Index Labels
# A. Renaming Row Index (Single Row)
# Scenario: Correcting an incorrect index label.
df = pd.DataFrame({
    'sales': [100, 200, 300]
}, index=['store1', 'store2', 'stor3'])

df = df.rename(index={'stor3': 'store3'})
print(df.index)

# B. Renaming All Index Labels (Using a Function)
# Scenario: Adding a prefix to all index labels.
df = df.rename(index=lambda x: 'prefix_' + x)
print(df.index)

df=df.rename(index=lambda x: x.replace('prefix_',''))
print(df.index)

df=df.rename(index=lambda x: f'prefix_{x}')
print(df.index)


# 3. Renaming Both Columns & Index at Once
# Scenario: Standardizing a DataFrame with inconsistent naming.
df = pd.DataFrame({
    'Sales_Amt': [100, 200, 300],
    'Qty': [5, 10, 15]
}, index=['st1', 'st2', 'st3'])

df = df.rename(
    columns={'Sales_Amt': 'sales', 'Qty': 'quantity'},
    index={'st1': 'store1', 'st2': 'store2', 'st3': 'store3'}
)
print(df)


# 4. Real-World Use Cases
# Case 1: Cleaning Survey Data (Inconsistent Column Names)
survey_data = pd.DataFrame({
    'Respondent ID': [1, 2, 3],
    'Age (Years)': [25, 30, 35],
    'Income ($)': [50000, 60000, 70000]
})

survey_data = survey_data.rename(columns={
    'Respondent ID': 'respondent_id',
    'Age (Years)': 'age',
    'Income ($)': 'income'
})
print(survey_data.columns)

# Case 2: Financial Data (Standardizing Column Names)
stock_data = pd.DataFrame({
    'Ticker': ['AAPL', 'GOOG', 'MSFT'],
    'Prc': [150, 2800, 300],
    'Vol': [1000000, 500000, 2000000]
})

stock_data = stock_data.rename(columns={
    'Prc': 'price',
    'Vol': 'volume'
})
print(stock_data.columns)
```
