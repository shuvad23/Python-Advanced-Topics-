```python
# Flattening a Nested Dictionary
# Convert a nested dictionary into a flat dictionary.
def flatten_dict(d,parent_key='',sep='_'):
  item=[]
  for key,value in d.items():
    new_key=f'{parent_key}{sep}{key}' if parent_key else key
    if isinstance(value,dict):
      item.extend(flatten_dict(value,new_key,sep=sep).items())
    else:
      item.append((new_key,value))
      # print(item)
  return dict(item)




nested_dict = {'a': 1, 'b': {'c': 2, 'd': {'e': 3}}}
flat_dict = flatten_dict(nested_dict)
print(flat_dict)
```

```python
from collections import defaultdict
import time

def is_anagram(list_1,list_2):
    start_time=time.time()
    # Create mapping for `list_1`
    mapping_1 = defaultdict(list)
    # mapping_1 stores (sorted anagram) -> list[anagrams] mapping for `list_1`
    for word in list_1:
        sorted_tuple = tuple(sorted(word)) # unique identifier of the anagram
        mapping_1[sorted_tuple].append(word)
        # `mapping_1[sorted_tuple]` stores all anagrams under the same identifier for `list_1`

    # Create mapping for `list_2`
    mapping_2 = defaultdict(list)
    # mapping_2 stores (sorted anagram) -> list[anagrams] mapping for `list_2`
    for word in list_2:
        sorted_tuple = tuple(sorted(word)) # unique identifier of the anagram
        mapping_2[sorted_tuple].append(word)
        # `mapping_2[sorted_tuple]` stores all anagrams under the same identifier for `list_2`

    # Intersect keys from mapping_1 and mapping_2 to get common sorted tuples
    # Every element in `common_tuples` is an anagram identifier that exists in both lists
    common_tuples = set(mapping_1.keys()) & set(mapping_2.keys())

    output = []
    for anagram_tuple in common_tuples:
        for word1 in mapping_1[anagram_tuple]:
            for word2 in mapping_2[anagram_tuple]:
                # Both word1 and word2 have the same anagram identifier, so are anagrams
                output.append((word1, word2))
    print(f'Time taken: {time.time()-start_time}')
    return output

list1 = ["listen", "triangle", "elbow", "state", "night", "heart", "earth", "study", "dusty", "save"]*100
list2 = ["silent", "integral", "beow", "taste", "thing", "hater", "hate", "duty", "dust", "vase"]*100
print(is_anagram(list1,list2))
```

```python
#  Character Count Approach (More Efficient)

from collections import defaultdict

def is_anagram(str1,str2):
  str1=str1.replace(' ','').lower()
  str2=str2.replace(' ','').lower()

  if(len(str1)!=len(str2)):
    return False

  count=defaultdict(int)

  for char in str1:
    count[char] +=1
  for char in str2:
    count[char]-=1
    if count[char] < 0:
      return False

  return True

str1="hello world"
str2="world hello"
is_anagram(str1,str2)


# Time Complexity: O(n)
# Space Complexity: O(1) - fixed size alphabet storage
```

## Collision Handling in Hash Tables
1. Chaining: In this method, each index (or bucket) in the array hosts a linked list of all key-value pairs that hash to the same index. When a collision occurs, we simply go to the collided index and append the new key-value pair to the existing linked list.

2. Open Addressing: Upon encountering a collision, the hash table searches for another free slot or index in the table (possibly the next available empty slot) and assigns that location to the new key-value pair. This approach requires a suitable probing strategy to ensure efficient use of table space.

## Time and Space Complexity Analysis for Hash Tables

Hash tables are renowned for their efficiency and speed when it comes to data storage and retrieval. They boast constant time complexity O(1) for the operations on key-value pairs - insertion, deletion, and retrieval. This efficiency comes from a good hash function, which allows for keys to be uniformly distributed across the table and accessed directly via their indices, eliminating the need to scan through unnecessary slots.

Although hash tables generally perform robustly, situations may arise where frequent collisions occur. Such situations could deteriorate the table's efficiency and extend the time complexity to a worst-case scenario of
O(n), where n is the number of keys hashing to the same index.

##Hashing in Python: Beginner to Advanced with Real-Life Examples
Hashing is a fundamental concept in computer science that converts data into a fixed-size value (hash) for efficient storage and retrieval. Python uses hashing in dictionaries, sets, and security applications.

##1. Beginner Level: Understanding Hashing

What is Hashing?
1. Hashing converts input data (like strings, numbers, or objects) into a fixed-size integer (hash value).

2. Used in dictionaries (dict) and sets (set) for fast lookups (O(1) average time complexity).

```python
# Hashing different data types
print(hash("hello"))       # Hash of a string
print(hash(42))            # Hash of an integer
print(hash((1, 2, 3)))     # Hash of a tuple (immutable)

# Unhashable types (mutable objects) will raise an error
# print(hash([1, 2, 3]))   # TypeError: unhashable type: 'list'

# Key Points:
#1.  Same input â†’ same hash (in the same Python session).
#2.  Used in dictionaries (dict) and sets (set) for fast lookups.
```

2. Intermediate Level: Hash Tables (Dictionaries)

  How Python Dictionaries Use Hashing
  Dictionaries (dict) use hashing to store key-value pairs efficiently.

```python
user_db={}

def register_user(username,password):
  user_db[username]=hash(password)

def login(username,password):
  if username in user_db and user_db[username]==hash(password):
    print("Login successful")
  else:
    print("Login failed")


# Register users
register_user("alice", "secure123")
register_user("bob", "python456")

# Login attempts
login("alice", "secure123")  # Success
login("bob", "wrongpass")    # Fails


# How it works:
#1.  hash(password) converts the password into an integer.

#2.  The dictionary (user_db) stores username: hashed_password.

#3.  During login, Python compares hashes instead of plaintext passwords.
```

3. Advanced Level: Cryptographic Hashing (hashlib)

  For security (e.g., password storage), Pythonâ€™s hashlib provides stronger hashing algorithms like SHA-256.

  Example: Secure Password Storage

```python
import hashlib

def hash_password(password):
  # Using SHA-256 with a salt (for security)
  salt="randomsalt123"
  hashed=hashlib.sha256((password+salt).encode()).hexdigest()
  print(hashed)
  return hashed

# Store hashed passwords (never store plaintext!)
user_db_secure ={
    "alice": hash_password("secure123"),
    "bob": hash_password("python456")
}


def secure_login(username,password):
  if username in user_db_secure and user_db_secure[username]==hash_password(password):
    print("Secure login successful!")
  else:
    print("Secure login failed!")

secure_login("alice","secure123") # Success
secure_login("bob","python456") # Fails


# Why hashlib is better than hash()?
# 1. hash() is not secure (deterministic, reversible in some cases).
# 2. hashlib uses cryptographic hashing (SHA-256, MD5, etc.) for security.
```

How Python's hash() Works Internally


```python
# Basic hashing examples
print(hash("Python"))          # 5577755419904001916 (may vary)
print(hash(3.14159))          # 326490430436040707
print(hash((1, 2, "three")))  # -2023806683280880616

# Interesting behavior with numeric values
print(hash(1) == hash(1.0))   # True - numerically equal

# Key Insights:

# 1. Integers hash to themselves (except -1 due to implementation)

# 2. Python salts string hashes at interpreter startup for security

# 3. Custom objects use memory address by default (unless __hash__ is overridden)
```

```python
# Custom Object Hashing

class User:
  def __init__(self,id,username):
    self.id=id
    self.username=username

  def __hash__(self):
    return hash((self.id,self.username))
  def __eq__(self, __o: object) -> bool:
    return (self.id,self.username) == (__o.id,__o.username)


user1=User(1,'Alice')
user2=User(1,'Alice')

print(hash(user1)==hash(user2))

print(hash(user1))
print(hash(user2))
```

Secure Password Storage Best Practices

```python
# Secure Password Storage Best Practices
import hashlib
import os
import binascii

def create_password_hash(password):
  salt=os.urandom(16)
  iterations=100000
  key=hashlib.pbkdf2_hmac(
      'sha256',
      password.encode('utf-8'),
      salt,
      iterations
  )
  return f'{iterations}:{binascii.hexlify(salt).decode()}:{binascii.hexlify(key).decode()}'
def verify_password(stored_hash,password):
  iterations,salt,key=stored_hash.split(":")
  new_key=hashlib.pbkdf2_hmac(
      'sha256',
      password.encode('utf-8'),
      binascii.unhexlify(salt),
      int(iterations)
  )
  return binascii.hexlify(new_key).decode() == key


# Usage
stored = create_password_hash("secure123")
print(stored)
print(verify_password(stored, "secure123"))  # True
print(verify_password(stored, "wrongpass"))  # False
```

4. Advanced Applications

```python
# Content-Addressable Storage System
import hashlib
class ContentStore:
  def __init__(self):
    self.storage={}
  def store(self,data):
    key=hashlib.sha256(data).hexdigest()
    self.storage[key]=data
    return key
  def retrieve(self,key):
    return self.storage.get(key)

store=ContentStore()
doc= b"Important document content"
doc_key=store.store(doc)
print(f"Stored document with key: {doc_key}")
print(f"Retrieved: {store.retrieve(doc_key)}")
```

```python
# Bloom Filter Implementation

# !pip install bitarray
# !pip install mmh3

from bitarray import bitarray
import mmh3 # MurmurHash for better distribution
class BloomFilter:
  def __init__(self,size,hash_count):
    self.size=size
    self.hash_count=hash_count
    self.bit_array=bitarray(size)
    self.bit_array.setall(0)

  def add(self,item):
    for seed in range(self.hash_count):
      index=mmh3.hash(item,seed)%self.size
      self.bit_array[index]=1

  def __contains__(self,item):
    for seed in range(self.hash_count):
      index=mmh3.hash(item,seed) %self.size
      if not self.bit_array[index]:
        return False
      return True
bf=BloomFilter(100,3)
bf.add("test@example.com")
print("test@example.com" in bf)
print("unknown@example.com" in bf)
bf.add("max@example.com")
print("max@example.com" in bf)
```

5. Performance Optimization Techniques

```python
# Custom Hashing for Speed
def fast_hash(items):
  h=2166136261
  for item in items:
    h=(h*16777619) ^ hash(item)
  return h

from timeit import timeit
data =tuple(range(1000))
print(timeit(lambda: hash(data),number=1000)) # Standard
print(timeit(lambda: fast_hash(data),number=1000)) # Custom
```

```python
# Dictionary View Optimizations

large_dict = {i: str(i) for i in range(10_000)}

# Slow iteration
def slow():
    for k in large_dict:
        pass

# Faster using views
def fast():
    for k in large_dict.keys():
        pass

print(timeit(slow, number=1000))
print(timeit(fast, number=1000))
```

6. Security Considerations

```python
# Hash Flooding Protection
# Python 3.3+ includes hash randomization to prevent denial-of-service attacks:

# 1. Observe different hashes across Python runs
print(hash("attack string"))  # Changes with each interpreter launch


# 2. Timing Attack Prevention
import hmac

def safe_compare(a,b):
  return hmac.compare_digest(a,b)

print(safe_compare("hello", "world"))  # False
print(safe_compare("hello", "hello"))  # True
```
